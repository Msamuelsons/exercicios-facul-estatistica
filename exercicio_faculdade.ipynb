## Exercicio 1

Uma indústria inspecionou 30 aparelhos fabricados por certa máquina, obtendo os

seguintes números de defeitos por aparelho.
<p>1 0 2 0 1 1 4 0 2 3 2 2 1 0 0</p>
<p>0 2 1 1 0 0 4 2 1 0 0 0 0 1 2</p>
- Elabore uma distribuição de frequência e calcule a frequência acumulada e relativa




```python
import pandas as pd
```

```python
data = [1, 0, 2, 0, 1, 1, 4, 0, 2, 3, 2, 2, 1, 0, 0,0, 2, 1, 1, 0, 0, 4, 2, 1, 0, 0, 0, 0,1, 2]
rol_data = sorted(data)
```

```python
print(rol_data, end='')
```

```python
df = pd.DataFrame(rol_data, columns= ['Valores'])
df
```

```python
df.value_counts()
```

```python
tabela_frequencia = df.groupby('Valores').size().reset_index(name='Frequência Absoluta')
```

```python
cont = 0
frequencia_absoluta = []
for x in range(0, len(tabela_frequencia['Frequência Absoluta'])):
    frequencia_absoluta.append(tabela_frequencia['Frequência Absoluta'][x])
    cont+=tabela_frequencia['Frequência Absoluta'][x]
frequencia_total = cont
```

```python
frequencia_acumulativa = []
frequencia_absoluta = frequencia_absoluta
acumulativo = 0
for x in range(0, len(tabela_frequencia['Frequência Absoluta'])):
    acumulativo+=tabela_frequencia['Frequência Absoluta'][x]
    frequencia_acumulativa.append(acumulativo)
frequencia_acumulativa
tabela_frequencia['Acumulatica'] = frequencia_acumulativa
```

```python
frequencia_relativa = []
for x in range(0, len(frequencia_absoluta)):
    frequencia_relativa.append(frequencia_absoluta[x] / frequencia_total)
```

```python
tabela_frequencia['Relativa'] = frequencia_relativa
```

```python
tabela_frequencia
```

## Exercicio 2

Em uma universidade, vinte alunos foram submetidos a um teste de nivelamento
obtendo os seguintes resultados:

<p>26 28 24 13 18</p>
<p>18 25 18 25 24</p>
<p>20 21 15 28 17</p>
<p>27 22 13 19 28</p>

1.  Elaborar distribuição de frequência por classe
2. Calcular fa e fr
3. Calcular o Pm de cada classe

```python
notas_alunos = [26, 28, 24, 13, 18, 18, 25, 18, 25, 24, 20, 21, 15, 28, 17,27, 22, 13, 19, 28]

notas = {'Notas': notas_alunos}

df_notas = pd.DataFrame(notas)
df_notas
```

```python
frequencia_absoluta_notas = df_notas['Notas'].value_counts()
frequencia_absoluta_notas
```

```python
tabela_frequencia_notas = df_notas.groupby('Notas').size().reset_index(name='Frequência Absoluta')
```

```python
tabela_frequencia_notas
```

```python
frequencia_total_notas = tabela_frequencia_notas['Frequência Absoluta'].sum()
frequencia_total_notas
```

```python
frequencia_relativa_notas = []

for x in range(0, len(tabela_frequencia_notas['Frequência Absoluta'])):
  frequencia_relativa_notas.append(tabela_frequencia_notas['Frequência Absoluta'][x] / frequencia_total_notas)
```

```python
frequencia_relativa_notas
```

```python
tabela_frequencia_notas['Relativa'] = frequencia_relativa_notas
```

```python
tabela_frequencia_notas
```

```python
valor_maximo = tabela_frequencia_notas['Frequência Absoluta'].max()
valor_minimo = tabela_frequencia_notas['Frequência Absoluta'].min()
```

```python
numero_de_classes = len(tabela_frequencia_notas['Frequência Absoluta'].unique())
numero_de_classes
```

```python
amplitude_classe = (valor_maximo - valor_minimo) / numero_de_classes
amplitude_classe
```
